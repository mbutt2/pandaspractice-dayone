```python
people = {
    'first':['Mehreen', 'Stephen', 'Namita', 'Rachel', 'Neal', 'Hannah'],
    'last': ['Butt', 'Castaneda', 'Manohar', 'Usry', 'Usry', 'Feutz'],
    'email': ['mehreen.butt@email.com', 'stephen.castaneda@email.com', 'namita.manohar@email.com', 'rachel.usry@email.com', 'neal.usry@email.com', 'hannah.feutz@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='last')
# sorting by last name alphabetically
```

```python
df.sort_values(by='last', ascending=False)
# sorting by last name in descending order using ascending = False 
```

```python
df.sort_values(by=['last', 'first'], ascending=False)
# didn't change much here, but sorted last name first in descending order, then sorted duplicates in descending order - which already was organized that way...R and then N
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
# sorts in descending and then the second argument in ascending...N --> R for the Usrys
```

```python
df
```

```python
df.sort_index()
# sorting in the way that the values were added
```

```python
df['first'].sort_values()
# sorting just the first name column
```
